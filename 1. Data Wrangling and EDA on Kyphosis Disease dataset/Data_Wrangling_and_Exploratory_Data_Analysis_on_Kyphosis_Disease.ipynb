### **In this project I will be performing basic Exploratory Data Analysis (EDA) on the Kyphosis Disease Dataset.**
### Kyphosis is an abnormally excessive concave curvature of the spine.

### **1. Importing the "Kyphosis.csv" file using Pandas**

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)


kyphosis_df = pd.read_csv('kyphosis.csv')
kyphosis_df
```

```python
kyphosis_df.info()
```

```python
kyphosis_df.describe()
```

```python
kyphosis_df.isnull().sum() #There are no null values in the Dataset
```

### **2. Converting 'Age' column from 'Months' to 'Years'**

```python
def months_to_years(age):
    return age/12

kyphosis_df['Age in Years'] = kyphosis_df['Age'].apply(months_to_years).round(1)

kyphosis_df
```

```python
kyphosis_df.head()
```

### **3. Seperating present and absent entries in the data and performing basic EDA**


```python

present_kyphosis_df = kyphosis_df[(kyphosis_df['Kyphosis']=='present')]
absent_kyphosis_df = kyphosis_df[(kyphosis_df['Kyphosis']=='absent')]
```

```python
present_kyphosis_df.describe()
```

```python
absent_kyphosis_df.describe()
```

Checking the features of oldest and youngest child in the study


```python
oldest_child = kyphosis_df[kyphosis_df['Age in Years'] == kyphosis_df['Age in Years'].max()]
print(oldest_child)
```

```python
youngest_child = kyphosis_df[kyphosis_df['Age in Years'] == kyphosis_df['Age in Years'].min()]
print(youngest_child)
```

### 4. Plotting Heatmap of the DataSet

```python
import matplotlib.pyplot as plt
import seaborn as sns
correlations = kyphosis_df.corr()

f,ax = plt.subplots(figsize = (5,5))
sns.heatmap(correlations, annot=True)
```
