### In this project I have visualized stock prices using Seaborn and Matplotlib.

4 stocks are considered including JP Morgan Chase (JP), Procter and Gamble (P&G) (PG), Apple (APPL) and United Airlines (UAL)

```python
# Importing necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing both datasets using Pandas

prices_df = pd.read_csv('stocks_daily_prices.csv')
returns_df = pd.read_csv('stocks_daily_returns.csv')
```

```python
# Plotting lineplots that display all 4 stocks daily prices on one single figure

prices_df.plot(x='Date', label='Stocks daily prices', linewidth=1, figsize =(14,6))
plt.ylabel('Prices')
plt.legend(loc = 'upper left')
plt.grid()
```

```python
# Plotting in multiple subplots
prices_df.plot(x='Date', linewidth=1, figsize = (14,6), subplots = True)
plt.ylabel('Prices')
plt.grid()
```

```python
#Plotting a scatterplot between Apple and JP Morgan Daily Rates

returns_df.plot.scatter('AAPL', 'JPM', grid = True, figsize = (8,8), color = 'hotpink', alpha = 0.5)
```

```python
# Using Seaborn to plot similar scatteeplot between AAPL and JPM daily returns

sns.scatterplot(x = 'AAPL', y = 'JPM', data = returns_df)
```

```python
# I decided to become bullish on AAPL and I allocated 70% of my assets in it. I also decided to equally divide the rest of my assets in other stocks (JPM, PG, and UAL).
# Using matplotlib im plotting a pie chart that shows these allocations.

values  = [10, 70, 10, 10]
colors  = ['g', 'r', 'y', 'b']
explode = [0, 0.2, 0, 0]
labels  = ['JP Morgan Chase (JPM)', 'APPLE (AAPL)', 'United Airlines (UAL)', 'P&G (PG)']

plt.figure(figsize = (10, 10))
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('STOCK PORTFOLIO')
plt.show()
```

```python
# Plotting HISTOGRAM for United Airlines and P&G returns. Also displaying mean and standard deviation for both stocks in figure
```

```python

plt.figure(figsize = (20, 15));

mu_UAL = round(returns_df['UAL'].mean(), 2)
sigma_UAL = round(returns_df['UAL'].std(), 2)


mu_PG = round(returns_df['PG'].mean(), 2)
sigma_PG = round(returns_df['PG'].std(), 2)


returns_df[['PG','UAL']].plot.hist(bins = 60, alpha = 0.7, figsize = (15, 8));
plt.grid();
plt.ylabel('Probability')


plt.title('Histogram: P&G mu=' + str(mu_PG) + ', P&G sigma=' +str(sigma_PG) + ', United Airlines mu=' +str(mu_UAL) + ', United Airlines sigma=' +str(sigma_UAL) );
```

```python
# Plotting a heatmap that shows correlations between stocks daily returns

sns.heatmap(returns_df.corr(), annot = True)
plt.figure(figsize = (20,20))
```

```python
# Plotting a 3d scatterplot using toolkits
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(18, 18))
ax = fig.add_subplot(111, projection = '3d')

x = returns_df['JPM'].tolist()
y = returns_df['AAPL'].tolist()
z = returns_df['UAL'].tolist()

ax.scatter(x, y, z, c = 'r', marker = 'o', s = 1000)

ax.set_xlabel('JP Morgan')
ax.set_ylabel('Apple')
ax.set_zlabel('United Airlines')
```
